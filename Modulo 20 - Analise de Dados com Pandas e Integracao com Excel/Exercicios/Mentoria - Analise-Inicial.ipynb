### Puxar base de dados

```python
import pandas as pd

df_exportacoes = pd.read_csv(r'Arquivos Base\Mentoria - exportacoes_franca.csv')

# df_exportacoes = df_exportacoes.loc[df_exportacoes['Economic Block'] == 'Europe' , :]
df_exportacoes = df_exportacoes[df_exportacoes['Economic Block'] == 'Europe']

df_exportacoes
```

### Informações Gerais

```python
print(df_exportacoes.info())
# display(df_exportacoes.describe())
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

```python
# criando uma formatação
def formatar(valor):
    valor_formatado = f'${valor:,.2f}'.replace(',', '_').replace('.', ',')
    valor_formatado = valor_formatado.replace('_', '.')
    return valor_formatado


exportacao_ano = df_exportacoes.groupby(['Year']).sum(numeric_only=True)[['US$ FOB']]
exportacao_ano.plot(kind='bar')

exportacao_ano['US$ FOB'] = exportacao_ano['US$ FOB'].map(formatar)
display(exportacao_ano)
```

```python
import plotly.express as px

exportacao_ano = exportacao_ano.reset_index()
grafico = px.line(exportacao_ano, x='Year', y='US$ FOB', title='Exportações para a França por Ano')
grafico.update_xaxes(type='category')
grafico.update_yaxes(rangemode='tozero')

grafico.show()
```

### Quais os produtos mais exportados ao longo de todo o período?

```python
produtos_mais_exportado = df_exportacoes.groupby('SH4 Description').sum(numeric_only=True)[['US$ FOB']].sort_values(by='US$ FOB', ascending=False)
produtos_mais_exportado['US$ FOB'] = produtos_mais_exportado['US$ FOB'].map(formatar)
produtos_mais_exportado
```

### Em 2020 qual cidade mais exportou para a França?

```python
df_2020 = df_exportacoes.loc[df_exportacoes['Year'] == 2020, :]
df_2020.groupby('City').sum(numeric_only=True)[['US$ FOB']].sort_values(by='US$ FOB', ascending=False)
```

### O que as 2 maiores cidades exportaram tanto?

```python
df_2020.loc[df_exportacoes['City'] == 'Duque de Caxias - RJ', :].groupby('SH4 Description')[['US$ FOB']].sum(numeric_only=True).sort_values(by='US$ FOB', ascending=False).map(formatar)
```
