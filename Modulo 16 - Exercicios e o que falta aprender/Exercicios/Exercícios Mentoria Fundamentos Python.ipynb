### 1- Imposto a pagar no Lucro Presumido

- 5% sobre faturamento de ISS (mensal)
- 0,65% de PIS sobre faturamento, (mensal)
- 3% de COFINS sobre faturmaneto, (mensal)
- 4.8% de IR (trimestral)
- 10% de IR Adicional sobre o que ultrapassar 20mil do faturamento (trimestral)
- CSLL: 2,88% sobre faturamento (trimestral)

#### *Você precisa inserir no sistema um dicionário no formato*:

```
resultado = {
    'mes': (faturamento, imposto_mensal, imposto_trimestral),
}
```

````python
faturamento = {
    'jan': 'R$ 95.141,98',
    'fev': 'R$ 95.425,16',
    'mar': 'R$ 89.716,31',
    'abr': 'R$ 78.459,99',
    'mai': 'R$ 71.087,28',
    'jun': 'R$ 83.911,06',
    'jul': 'R$ 56.467,26',
    'ago': 'R$ 88.513,58',
    'set': 'R$ 66.552,49',
    'out': 'R$ 80.164,07',
    'nov': 'R$ 66.964,33',
    'dez': 'R$ 71.525,25',
}


def calcula_imposto_mensal(dicionario):

    def calcula_iss(faturamento):
        valor_iss = 0.05 * faturamento
        return valor_iss
    
    def calcula_pis(faturamento):
        valor_pis = 0.0065 * faturamento
        return valor_pis
    
    def calcula_cofins(faturamento):
        valor_cofins = 0.03 * faturamento
        return valor_cofins

    imposto_mensal = {}

    for mes, faturamento in dicionario.items():
        faturamento = float(faturamento[3:].replace('.', '').replace(',', '.'))
        imposto_mensal[mes] = round(calcula_iss(faturamento) + calcula_cofins(faturamento) + calcula_pis(faturamento), 2)

    return imposto_mensal

def calcula_imposto_trimestral(dicionario):

    def calcula_ir(faturamento):
        valor_ir = 0.048 * faturamento

        if faturamento > 20_000:
            valor_ir += (faturamento - 20_000) * 0.1

        return valor_ir
    
    def calcula_csll(faturamento):
        valor_csll = 0.0288 * faturamento

        return valor_csll
    
    imposto_trimestral = {}

    for mes, faturamento in dicionario.items():
        faturamento = float(faturamento[3:].replace('.', '').replace(',', '.'))

        imposto_trimestral[mes] = round(calcula_ir(faturamento) + calcula_csll(faturamento), 2)
    
    return imposto_trimestral
    
def main(dicionario, imposto_mensal, imposto_trimestral):
    resultado = {}

    for index, mes_faturamento in enumerate(dicionario.items()):
        mes, faturamento = mes_faturamento
        resultado[mes] = (faturamento, f'R$ {imposto_mensal[index]:,.2f}', f'R$ {imposto_trimestral[index]:,.2f}')
    
    return resultado


imposto_mensal = list(calcula_imposto_mensal(faturamento).values())
imposto_trimestral = list(calcula_imposto_trimestral(faturamento).values())

resultado = main(faturamento, imposto_mensal, imposto_trimestral)
print(resultado)
````

### 3 - Extração dos links de Download de Vídeos do Vimeo

- Precisamos pegar os links em 1080p, 720p e 540p para importar os vídeos para uma nova plataforma

````python
# para pegar o dicionario do vimeo, use:
from dic import dicionario_vimeo

videos = []

for key, value in dicionario_vimeo.items():
    if key == 'data':
        for v in value:
            infos = {'uri': '', 'nome': '', 'duracao': '', 'link540p': 'NÃO TEM', 'link720p': 'NÃO TEM', 'link1080p': 'NÃO TEM'}

            uri = v['uri']
            infos['uri'] = uri

            nome = v['name']
            infos['nome'] = nome

            duracao = v['duration']
            infos['duracao'] = duracao
            
            for cont in range(len(v['download'])):
                    if v['download'][cont]['rendition'] == '540p':
                        link540p = v['download'][cont]['link']
                        infos['link540p'] = link540p

                    elif v['download'][cont]['rendition'] == '720p':
                        link720p = v['download'][cont]['link']
                        infos['link720p'] = link720p

                    elif v['download'][cont]['rendition'] == '1080p':
                        link1080p = v['download'][cont]['link']
                        infos['link1080p'] = link1080p

            videos.append(infos)

print(videos)

# você precisa chegar em:
# videos = [
#     {'uri': video_uri, 'nome': nome_video, 'duracao': duracao, 'link540p': link540p, 'link720p': link720p, 'link1080p': link1080p},
# ]
````
