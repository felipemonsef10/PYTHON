```python
from pathlib import Path
import pandas as pd
import win32com.client as win32

outlook = win32.Dispatch('outlook.application')
arquivos_excel = Path('Arquivos Excel').iterdir()
```

```python
df_emails = pd.read_excel(r"Arquivos Excel\Enviar E-mails.xlsx")
df_emails['E-mail'] = ['pikatofly17@gmail.com' for c in range(len(df_emails['E-mail']))]


for i, gerente in enumerate(df_emails['Gerente']):
    mail = outlook.CreateItem(0)
    mail.Subject = 'Relatório'

    mensagem = f"""<p>Boa tarde {gerente}!</p>
<p>Segue em anexo o relatório referente ao report diário</p>
<p>de seu setor.</p>
    """
    mail.To = df_emails['E-mail'].iloc[i]
    mail.HTMLBody = mensagem

    for arquivo in Path("Arquivos Excel").iterdir():
        if df_emails['Relatório'].iloc[i] in arquivo.name:
            attachment = str(arquivo.absolute())
            break
    mail.Attachments.Add(attachment)

    mail.Send()
```
